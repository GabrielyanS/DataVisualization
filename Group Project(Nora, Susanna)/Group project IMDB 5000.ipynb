```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('movie_metadata.csv')
```

```python
data = data.dropna()
data.head()
```

```python
data["genres"] = data["genres"].str.split(pat="|").str[0]
```

```python
data_genres = data.groupby("genres").size()
data_genres = data_genres.reset_index()
```

```python
data_genres = data_genres.sort_values(by = 0, ascending = False)
```

We want to find the most popular genres. Thats why we have ploted the bar plot, which you can see below. In the dash application we will plot this graph for every country(using callbacks). (Nora)

```python
sns.barplot(data = data_genres, x = 0, y = 'genres')
```

```python
data["profit"] = data["gross"] - data["budget"]
```

```python
data["profit"]
```

```python
data_score = data[data["imdb_score"]>8]
data_score = data_score.sort_values("imdb_score", ascending = False)
data_score
```

```python
data_score.head(20)
```

Two linecharts, one for score and the second one for profit, to know if the higher score movie has higher profit. In dash this will be 2 graphs in one raw, and each graph will be interactive and user could see linecharts for each year.(Susanna)

```python
sns.lineplot(data = data_score.head(20) , x = "movie_title", y = "imdb_score" )
```

```python
sns.lineplot(data = data_score.head(20), y = "profit" , x = "movie_title" )
```

We pick the most popular genres and plot the boxplot of IMDB scores. In dash application we will see this graph for each country and in the upcoming graph (bar plot) we will see the mean score for that selected country for each genre. So these last two graphs will be connected.(Nora)

```python
genres_top_5 = ["Comedy","Action","Drama","Adventure","Crime"]
data_geners_top_5 = data[data.genres.isin(genres_top_5)]
sns.boxplot(data = data_geners_top_5, x = "genres" , y = "imdb_score")
plt.xticks(rotation = 90)
```

```python
data_geners_top_5.groupby(["country","genres"]).mean()[['imdb_score']]
```

```python
data_geners_top_5.groupby(["country"]).size().sort_values(ascending = False)
```

```python
top_10_countries = ["USA","UK","France","Canada", "Germany","Australia","China","Spain","India","Japan" ]
genres_top_5 = ["Comedy","Action","Drama","Adventure","Crime"]
data_geners_top_5 = data[data.genres.isin(genres_top_5)]
data_genres_countries = data_geners_top_5[data_geners_top_5.country.isin(top_10_countries)]
```

```python
plt.figure(figsize = (26,25))
data_genres_countries.groupby(["country","genres"]).mean()[['imdb_score']].unstack().plot.bar()
```
